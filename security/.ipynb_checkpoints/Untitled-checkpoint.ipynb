```python
#Imports
import hashlib as md5_hash

#System starting function
def start_up():
    while(True):
        print("\nWelcome to Hospital Management System")
        print("1 - Login to the system")
        print("2 - New Patient sign up")
        print("3 - Exit\n")
        inputNumber = input("Enter number : ")
        if(inputNumber == "1"):
            system_login()
        elif(inputNumber == "2"):
            patient_signup()
        elif(inputNumber == "3"):
            break
        else:
            print("Invalid Input")

#Add patient to the configuration file
def patient_signup():
    user_name = input("\nEnter Username : ")
    password = md5_hash.md5(input("Enter a password : ").encode("utf-8")).hexdigest()
    user_type = "patient"
    user_privilage = "patient"
    with open("configuration.txt", "a") as file:
        file.write(user_name + ",#" + password + ",#" + user_type + ",#" + user_privilage + "\n")
    print("Successfully registered\n")
    
#System login function
def system_login():
    user_name = input("\nEnter Username : ")
    password = md5_hash.md5(input("Enter a password : ").encode("utf-8")).hexdigest()
    user_credentials = valdate_user(user_name,password)
    if(user_credentials == False):
        print("User not availabe\n")
    else:
        user_type = user_credentials[1]
        if(user_type=="patient"):
            read_patient_record(user_credentials[0],user_credentials[2])
        elif(user_type == "staff"):
            staff_function(user_credentials)
        else:
            print("Invalid User\n")
        
# FUnction for validate user name and password. Read from Configuratio file  
def valdate_user(u_name,password):
    is_user_exists = False
    user_credentials = []
    with open("configuration.txt", "r") as file:
        user_configuration_list = []
        for line in file:
            user = list(line.rstrip("\n").split(",#"))
            if(u_name == user[0] and password == user[1]):
                is_user_exists = True
                user_credentials = [user[0],user[2],user[3]]
                break
    if(is_user_exists):
        return user_credentials
    else:
        return False
    
def read_patient_record(user_name, category):
    if(category == "patient" or category == "doctor" or category == "nurse" 
       or category == "lab_tester"):
        with open("data.txt", "r") as file:
            for line in file:
                user_report = list(line.rstrip("\n").split(",#"))
                if(user_name == user_report[1]):
                    if(user_report[0] == "doc_rec"):
                        print("\nUser name : "+ user_report[1])
                        print("Sickness Details : "+ user_report[2])
                        print("Drug Prescription : "+ user_report[3])
                        print("Doctor Name : "+ user_report[4])
                        print()
                    elif(user_report[0] == "lab_rec"):
                        print("\nUser name : "+ user_report[1])
                        print("Lab Details : "+ user_report[2])
                        print("Laboratory : "+ user_report[3])
                        print()
    else:
        print("Access Denied")

# Give access to different function according to privilage
def staff_function(user_credentials):
    if(user_credentials[2] == "admin"):
        staff_registration(user_credentials[2])
    elif(user_credentials[2] == "doctor"):
        doctor_function(user_credentials)
    elif(user_credentials[2] == "nurse"):
        nurse_function(user_credentials)
    elif(user_credentials[2] == "lab_tester"):
        lab_tester_function(user_credentials)
    else:
        print("Invalid User")

# Staff register by admin. Only admin get access to staff registration
def staff_registration(user_category):
    if(user_category == "admin"):
        repeat = True
        while(repeat):
            print("\nStaff Registration\n")
            user_name = input("Enter member name : ")
            password = md5_hash.md5(input("Enter a password : ").encode("utf-8")).hexdigest()
            user_type = "staff"
            print("User privilage \n 1-doctor \n 2-nurse \n 3-lab_tester \n")
            privilage_number = input("Enter privilage number : ")
            privilage = "Invalid"
            if(privilage_number == "1"):
                privilage = "doctor"
            elif(privilage_number == "2"):
                privilage = "nurse"
            elif(privilage_number == "3"):
                privilage = "lab_tester"
            if(privilage != "Invalid"):
                with open("configuration.txt", "a") as file:
                    file.write(user_name + ",#" + password + ",#" + user_type 
                               + ",#" + privilage + "\n")
                print("Registration Successfull\n")
            else:
                print("Registration Failed\n")
            inp = input("Press 1 for exit, press 2 for register again: ")
            if(inp != "2"):
                repeat = False
            
def doctor_function(user_credentials):
    repeat = True
    while(repeat):
        print("\nDoctor Function \n 1-Read Patient record \n 2-Write record \n")
        user_action = input("Enter number : ")
        if(user_action == "1"):
            user_name = input("Enter patient user name : ")
            read_patient_record(user_name,user_credentials[2])
        elif(user_action == "2"):
            write_dataFile(user_credentials)
        inp = input("Press 1 for exit, press 2 for repeat: ")
        if(inp != "2"):
            repeat = False
            
# Lab tester access data file and write to it
def lab_tester_function(user_credentials):
    repeat = True
    while(repeat):
        print("\nLab Tester Function")
        write_dataFile(user_credentials)
        inp = input("Press 1 for exit, press 2 for repeat: ")
        if(inp != "2"):
            repeat = False

# Nurse can read data file
def nurse_function(user_credentials):
    repeat = True
    while(repeat):
        print("\nNurse Function - Read Patient Record")
        user_name = input("Enter patient user name : ")
        read_patient_record(user_name,user_credentials[2])
        inp = input("Press 1 for exit, press 2 for repeat: ")
        if(inp != "2"):
            repeat = False

# Write to data file according to user previlage 
def write_dataFile(user_credentials):
    if(user_credentials[2] == "doctor"):
        user_name = input("Enter patient user name : ")
        sickness = input("Enter sickness details : ")
        drug = input("Enter drug details : ")
        doc_name = user_credentials[0]
        rec_type = "doc_rec"
        with open("data.txt", "a") as file:
            file.write(rec_type + ",#" + user_name + ",#" + sickness 
                           + ",#" + drug + ",#" + doc_name + "\n")
    elif(user_credentials[2] == "lab_tester"):
        rec_type = "lab_rec"
        user_name = input("Enter patient user name : ")
        lab_rec = input("Enter Lab report : ")
        lab_name = user_credentials[0]
        with open("data.txt", "a") as file:
            file.write(rec_type + ",#" + user_name + ",#" + lab_rec  + ",#" + lab_name + "\n")
    print("Data added\n")
    
        
start_up()    
```

```python
import csv

#input_file  = input("Enter Input file Name :")
#output_file = input("Enter Output File Name :")

input_file = "test-input-4.csv"
output_file = "output.csv"
data_list= []
room_list = []
with open(input_file, newline='') as data:
    reader = csv.reader(data, delimiter='\t')
    for row in reader:
        data_list.append([data_set.strip() for data_set in list(filter(None, row[0].split(',')))])

room_list = data_list.pop(-1)
subject_data_dict = {}

for sub in data_list:
    subject_data_dict[sub[0]] = sub[1:]



def calc_slot_conflict(subject,slot,subject_data):
    csp_copy = subject_data.copy()
    csp_copy.pop(subject)
    sub_status = subject_data[subject][0]
    cons_val = 0;
    Conflict_with_optional = 0
    
    for sub in csp_copy.keys():
        for time in csp_copy[sub][1:]:
            if(slot==time):
                if(subject_data[subject][0] == 'o' and csp_copy[sub][0]=='o'):
                    cons_val+=0.5
                    Conflict_with_optional+=1
                else:
                    cons_val+=1
    return(cons_val,Conflict_with_optional)
    
def check_available(slot,sub_time_assign):
    if(slot not in sub_time_assign.values()):
        return True
    else:
        optional_conflicts = 0
        for key in sub_time_assign:
            if(sub_time_assign[key]==slot) and key[1]=='o':
                optional_conflicts+=1
        if(optional_conflicts > 0) and optional_conflicts<len(room_list):
            return True
        else:
            return False
    
def select_time_slots(subject, sub_time_assign, subject_data):
    slot_list = subject_data[subject][1:]
    slot_conflict_dict = {}
    for slot in slot_list:
        if(check_available(slot,sub_time_assign)):
            slot_conflict_dict[slot] = calc_slot_conflict(subject,slot,subject_data)
    slot_conflict_dict = dict(sorted(slot_conflict_dict.items(), key=lambda item: item[1]))
    return list(slot_conflict_dict.keys())
    
def select_least_slot_subjects(sub_time_assign,subject_data):
    unassigned_subjects = [key[0] for key in sub_time_assign.keys() if sub_time_assign[key]==None]
    sublist_least_slot = []
    min_value = float('inf')
    for sub in unassigned_subjects:
        if(min_value>len(subject_data[sub])):
            min_value = len(subject_data[sub])
    for sub in unassigned_subjects:
        if(min_value==len(subject_data[sub])):
            sublist_least_slot.append(sub)
    return sublist_least_slot
        
def calc_contrained_value(subject,subject_data):
    csp_copy = subject_data.copy()
    csp_copy.pop(subject)
    sub_status = subject_data[subject][0]
    cons_val = 0;
    Conflict_with_optional = 0
    for time_list in csp_copy.values():
        for num in range(1,len(time_list)):
            if(time_list[num] in subject_data[subject]):
                if(time_list[0] =='o' and sub_status=='o'):
                    cons_val+=0.5
                    Conflict_with_optional+=1
                else:
                    cons_val+=1
    return (cons_val,Conflict_with_optional)

def update_timeslot_domain(subject_data,time,state):
    if(state=='c'):
        for key in subject_data.keys():
            if time in subject_data[key]:
                subject_data[key].remove(time)
    return subject_data
    
def select_unassigned_variable(sub_time_assign, subject_data):
    sublist_min_slots = select_least_slot_subjects(sub_time_assign,subject_data)
    lower = -float('inf')
    upper = float('inf')
    selected_subject = ''
    for sub in sublist_min_slots:
        cons_value,conflicts = calc_contrained_value(sub,subject_data)
        if(cons_value>lower):
            lower,upper = cons_value,conflicts
            selected_subject = sub
        elif(cons_value==lower and conflicts<upper):
            lower,upper = cons_value,conflicts
            selected_subject = sub
    return selected_subject
        
    
def recursive_backtracking(sub_time_assign,subject_data):
    complete = False if None in sub_time_assign.values() else True
    if(complete==True):
        return True;
    else:
        subject = select_unassigned_variable(sub_time_assign,subject_data)
        time_slots = select_time_slots(subject, sub_time_assign, subject_data)
        for time in time_slots:
            sub_time_assign[(subject,subject_data[subject][0])] = time
            #change 
            subject_data_new = update_timeslot_domain(subject_data.copy(),time,subject_data[subject][0])
            new_result = recursive_backtracking(sub_time_assign, subject_data_new)
            if(new_result):
                return True
            else:
                sub_time_assign[(subject,subject_data[subject][0])] = None
    return False

def backtracking_search(subject_data):
    sub_time_assign = {}
    for sub in subject_data:
        sub_time_assign[(sub,subject_data[sub][0])] = None
    result = recursive_backtracking(sub_time_assign,subject_data)
    OuputFunction(sub_time_assign,result)
    
def OuputFunction(sub_time_assign,result):
    if(result==True):
        final_time_slot = list(sub_time_assign.values())
        slot_count = dict( (val,final_time_slot.count(val)) for val in set(final_time_slot))
        
    else:
        with open(output_file, 'w') as output:
            fieldnames = ['no result found']
            writer = output.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
    
backtracking_search(subject_data_dict)


```

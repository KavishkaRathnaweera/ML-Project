```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
3.29e7
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
# X,y should be dataframes | x=[1,2,3] not accepted. x=[[1,2,3]] is accepted.
```

Slope coefficient:

```python
regression.coef_    # theta_1
```

```python
#Intercept
regression.intercept_
```

```python
%matplotlib inline    

#export along python file

plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)

# Adding the regression line here:
plt.plot(X, regression.predict(X), color='red', linewidth=3)

#decoration
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.text(x=0,y=2,s='sample text',fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)

plt.style.use('ggplot')
#plt.style.use('classic')

plt.show()
```

```python
#Getting r square from Regression
regression.score(X, y)
```
